```python
#euclidean distance 
from math import*
 
def euclidean_distance(x,y):
    return sqrt(sum(pow(a-b,2) for a, b in zip(x, y)))
print(euclidean_distance([0,3,4,5],[7,6,3,-1]))
```

```python
#minkowski
from math import *
from decimal import Decimal

def nth_root(value, n_root):
    root_val = 1/float(n_root)
    return round(Decimal(value) ** Decimal(root_val), 3)

def minkowski(x, y, p_val):
    return nth_root(sum(pow(abs(a-b), p_val) for a,b in zip(x,y)), p_val)

print(minkowski([0, 3, 4, 5], [7, 6, 3, -1], 3))
    
```

```python
from math import*
 
def square_rooted(x):
    return round(sqrt(sum([a*a for a in x])),3)
 
def cosine_similarity(x,y):
    numerator = sum(a*b for a,b in zip(x,y))
    denominator = square_rooted(x)*square_rooted(y)
    return round(numerator/float(denominator),3)
print(cosine_similarity([5, 2, 3], [4, 1, 6]))
```

```python
from math import *

def cosine_sim(x, y):
    num = sum(a*b for a, b in zip(x, y))
    enovx = sqrt(sum([a*a for a in x]))
    enovy = sqrt(sum([a*a for a in y]))
    denom = enovx*enovy
    return (num/denom)
print(cosine_sim([5, 2, 3], [4, 1, 6]))
```
